```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
spotify_filepath = "data/spotify.csv"

# Read the file into a variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date")
spotify_data.sample(5)
```

>  Best to show trends over a period of time

### Line Charts


- Plot the data

```python
plt.figure(figsize=(14,6))

sns.lineplot(data=spotify_data)
```

```python
# convrt Date to Data time to fix x axis 

import datetime
spotify_data.index = pd.to_datetime(spotify_data.index)
spotify_data.info()
```

- Plot a subset of the data


```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))


# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You", dashes=False, marker="o")

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=spotify_data['Despacito'], label="Despacito",  dashes=False, marker="o")
```

 > Relationship: to understand relationships between variables 

### Bar Chart 

```python
plt.figure(figsize=(14,6))

sns.barplot(x = spotify_data.index.month , y= spotify_data['Shape of You'], color = 'pink')

```

### Heatmaps

```python

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=spotify_data.corr()
            , annot=True)
```

### Scatter plots

```python
sns.scatterplot(x= spotify_data.index.month , 
                y=spotify_data['Shape of You'])
```

- Color-coded scatter plots

```python
sns.scatterplot(x= spotify_data.index.month , 
                y=spotify_data['Shape of You'],
               hue = spotify_data.index.month)
```

### Regression line
To double-check the strength of this relationship, you might like to add a regression line, 

```python
sns.regplot(x= spotify_data.index.month , 
                y=spotify_data['Shape of You'])
```

### Histograms

```python
spotify_data.head()
```

```python

sns.distplot(a = spotify_data['Shape of You'], kde=False)
```

- 2D histogram

```python
sns.distplot(a = spotify_data['Shape of You'], kde=False, label="Shape of You")

sns.distplot(a = spotify_data['Despacito'], kde=False, label= "Despacito")

```

### KDE plots 

```python
sns.kdeplot(data = spotify_data['Shape of You'] , shade=True)

```

```python
sns.kdeplot(data = spotify_data['Shape of You'] , shade=True, label="Shape of You")

sns.kdeplot(data = spotify_data['Despacito'] , shade=True, label = "Despacito")

```

- 2D KDE plots

```python
sns.jointplot(x=spotify_data.index.month, 
              y=spotify_data['Shape of You'], kind="kde")
# kind : { "scatter" | "reg" | "resid" | "kde" | "hex" }
```
