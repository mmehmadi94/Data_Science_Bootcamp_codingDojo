Use Numpy to Solve the Systems of Equations

```python
import numpy as np
```

3x + 5y = 6

7x - 5y = 9

What is value of x and y ?

```python
A = np.array([[3,5],[7,-5]])
b = np.array([6,9])
x = np.linalg.solve(A,b)
x
```

Answer: x = 1.5 , y = 0.3
